```python
import requests
import time
import matplotlib.pyplot as plotty

# latty and longy and their evil cousin timey
latty=[]
longy=[]
timey=[]

for i in range(100):
    res=requests.get("https://api.wheretheiss.at/v1/satellites/25544")
    if res.status_code == 200:
        data = res.json()
        lat = data["latitude"]
        lon = data["longitude"]
        timestamp = data ["timestamp"]
        
        latty.append(lat)
        longy.append(lon)
        timey.append(timestamp)
        print(f"{i+1}: lat={lat:.2f} lon={lon:.2f}")
    time.sleep(10)

plotty.plot(longy, latty, 'o-', color='pink', markersize=5)
plotty.title("ISS Path (Last ~17 minutes)")
plotty.xlabel("Longitude")
plotty.ylabel("Latitude")
plotty.grid(True)
plotty.xlim(-180, 180)
plotty.ylim(-90, 90)
plotty.show()
```
